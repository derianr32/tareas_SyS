## 游냀 Simulaci칩n 2.4: Convoluci칩n Discreta

```python
import numpy as np

# --- Parte 1: Convoluci칩n con h[n] ---

print("--- PARTE 1: Convoluci칩n con h[n] ---")

# 1. Definir las secuencias de valores
# x[n] = {-15, 5, -3*, 0, 5, 7, -1}
# h[n] = {1, -2, 0*, 1, -2}
x_vals = np.array([-15, 5, -3, 0, 5, 7, -1])
h1_vals = np.array([1, -2, 0, 1, -2])

# 2. Definir los 칤ndices de n=0 para cada se침al
# En x_vals, -3 es el 3er elemento (칤ndice 2)
n_x0_idx = 2
# En h1_vals, 0 es el 3er elemento (칤ndice 2)
n_h1_0_idx = 2

# 3. Calcular la convoluci칩n
# 'full' calcula la convoluci칩n completa
y1_vals = np.convolve(x_vals, h1_vals, mode='full')

# 4. Calcular el 칤ndice n=0 de la salida
# n_y0 = n_x0 + n_h0. En 칤ndices de array:
n_y1_0_idx = n_x0_idx + n_h1_0_idx

print(f"Valores de y[n]: {y1_vals}")
print(f"El 칤ndice n=0 de y[n] est치 en la posici칩n: {n_y1_0_idx}")
print(f"El valor de y[0] es: {y1_vals[n_y1_0_idx]}")
print("Resultado y[n]: {-15, 35, -13, -9, 40* (n=0), -16, -4, -1, 5, -14, 2}\n")


# --- Parte 2: Convoluci칩n con Respuesta al Escal칩n s[n] ---

print("--- PARTE 2: Convoluci칩n con s[n] ---")

# 1. Definir la respuesta al escal칩n
# s[n] = {-1, 6, -10, 3*, 1, -10, 2, 5}
s_vals = np.array([-1, 6, -10, 3, 1, -10, 2, 5])
# El 칤ndice n=0 (valor 3) es el 4to elemento (칤ndice 3)
n_s0_idx = 3

# 2. Calcular h_nueva[n] = s[n] - s[n-1]
# Asumimos s[-inf] = 0.
# h[n] = s[n] - s[n-1]
# s_n_minus_1 se crea desplazando s_vals y rellenando con 0 al inicio
s_n_minus_1 = np.concatenate(([0], s_vals[:-1]))
h_main = s_vals - s_n_minus_1
# Tambi칠n necesitamos h[n] para el 칰ltimo punto: h[5] = s[5] - s[4] = 0 - 5 = -5
h_last = 0 - s_vals[-1]
h2_vals = np.concatenate((h_main, [h_last]))

# El 칤ndice n=0 de h_nueva es el mismo que el de s[n]
n_h2_0_idx = n_s0_idx # 칈ndice 3

print(f"Valores h_nueva[n]: {h2_vals}")
print(f"El valor h_nueva[0] es {h2_vals[n_h2_0_idx]} (es el 13)\n")

# 3. Calcular la nueva convoluci칩n: y_nueva = x * h_nueva
y2_vals = np.convolve(x_vals, h2_vals, mode='full')

# 4. Calcular el 칤ndice n=0 de la salida
n_y2_0_idx = n_x0_idx + n_h2_0_idx # 2 + 3 = 5

print(f"Valores de y_nueva[n]: {y2_vals}")
print(f"Longitud de y_nueva[n]: {len(y2_vals)} (7+9-1 = 15)")
print(f"El 칤ndice n=0 de y_nueva[n] est치 en la posici칩n: {n_y2_0_idx}")
print(f"El valor de y_nueva[0] es: {y2_vals[n_y2_0_idx]}")
```
